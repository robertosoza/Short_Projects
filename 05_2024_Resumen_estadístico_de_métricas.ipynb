```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from the spreadsheet
data1 = pd.read_excel('/2024.05.08 I-501 - 4.05.P01 Índ.Proces.Asuntos Antes del Plazo (iPA).xlsx', header=None, usecols='B', skiprows=1, nrows=28, names=['Valor real'])
data2 = pd.read_excel('/2024.05.08 I-502 - 4.05.P01 Índ. Registro Tareas dentro Plazo (iART).xlsx', header=None, usecols='B', skiprows=1, nrows=28, names=['Valor real'])
data3 = pd.read_excel('/2024.05.08 I-559 - 4.05.P01 Índice Archivo Op. Mej. en 6 meses (iAOP).xlsx', header=None, usecols='B', skiprows=1, nrows=20, names=['Valor real'])
data4 = pd.read_excel('/2024.05.08 I-489 - 4.05.P01 Índice Archivo Op. Mej (iAOM).xlsx', header=None, usecols='B', skiprows=1, nrows=28, names=['Valor real'])
```

```python
# Generate descriptive statistics
summary1 = data1.describe()
summary2 = data2.describe()
summary3 = data3.describe()
summary4 = data4.describe()

# Plot histogram
plt.figure(figsize=(20, 15))

plt.subplot(3, 4, 1)  # Create a subplot for the histogram
plt.hist(data1['Valor real'], bins=10, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histograma iPA-Procesamiento de Asuntos Antes del Plazo')
plt.xlabel('Valor Real')
plt.ylabel('Frecuencia')

plt.subplot(3, 4, 5)  # Create a subplot for the box plot
plt.boxplot(data1['Valor real'], vert=False)
plt.title('Box Plot')
plt.xlabel('Valor Real')

plt.subplot(3, 4, 2)  # Create a subplot for the histogram
plt.hist(data2['Valor real'], bins=10, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histograma iART-Registro Tareas dentro Plazo')
plt.xlabel('Valor Real')
plt.ylabel('Frecuencia')

plt.subplot(3, 4, 6)  # Create a subplot for the box plot
plt.boxplot(data2['Valor real'], vert=False)
plt.title('Box Plot')
plt.xlabel('Valor Real')

plt.subplot(3, 4, 3)  # Create a subplot for the histogram
plt.hist(data3['Valor real'], bins=10, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histograma iAOP-Archivo Op. Mejora en 6 m')
plt.xlabel('Valor Real')
plt.ylabel('Frecuencia')

plt.subplot(3, 4, 7)  # Create a subplot for the box plot
plt.boxplot(data3['Valor real'], vert=False)
plt.title('Box Plot')
plt.xlabel('Valor Real')

plt.subplot(3, 4, 4)  # Create a subplot for the histogram
plt.hist(data4['Valor real'], bins=10, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histograma iAOM-Archivo Op. Mejora')
plt.xlabel('Valor Real')
plt.ylabel('Frecuencia')

plt.subplot(3, 4, 8)  # Create a subplot for the box plot
plt.boxplot(data4['Valor real'], vert=False)
plt.title('Box Plot')
plt.xlabel('Valor Real')

summary1.loc['mean'] = summary1.loc['mean'].round(2)
summary1.loc['std'] = summary1.loc['std'].round(2)
summary1.loc['25%'] = summary1.loc['25%'].round(2)
summary4.loc['mean'] = summary4.loc['mean'].round(2)
summary2.loc['std'] = summary2.loc['std'].round(2)
summary3.loc['std'] = summary3.loc['std'].round(2)
summary4.loc['std'] = summary4.loc['std'].round(2)
summary2.loc['25%'] = summary2.loc['25%'].round(2)

plt.subplot(3, 4, 9)  # Create a subplot for the summary
plt.table(cellText=summary1.values,
          rowLabels=summary1.index,
          colLabels=summary1.columns,
          loc='center')
plt.axis('off')

plt.subplot(3, 4, 10)  # Create a subplot for the summary
plt.table(cellText=summary2.values,
          rowLabels=summary2.index,
          colLabels=summary2.columns,
          loc='center')
plt.axis('off')

plt.subplot(3, 4, 11)  # Create a subplot for the summary
plt.table(cellText=summary3.values,
          rowLabels=summary3.index,
          colLabels=summary3.columns,
          loc='center')
plt.axis('off')

plt.subplot(3, 4, 12)  # Create a subplot for the summary
plt.table(cellText=summary4.values,
          rowLabels=summary4.index,
          colLabels=summary4.columns,
          loc='center')
plt.axis('off')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# Read data from the spreadsheet
data5 = pd.read_excel('/2024.05.08 I-503 - 4.05.P01 EFICACIA PROCESO GESTIÓN OP. MEJORA.xlsx', header=None, usecols='B', skiprows=1, nrows=28, names=['Valor real'])
data6 = pd.read_excel('/2024.05.08 I-503 - 4.05.P01 EFICACIA PROCESO GESTIÓN OP. MEJORA.xlsx', header=None, usecols='A', skiprows=1, nrows=28, names=['Mes'])
```

```python
# Generate descriptive statistics
summary5 = data5.describe()

# Plot histogram
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)  # Create a subplot for the histogram
plt.hist(data5['Valor real'], bins=10, edgecolor='black')  # Adjust the number of bins as needed
plt.title('Histograma Eficacia Proceso GOM')
plt.xlabel('Valor Real')
plt.ylabel('Frecuencia')

plt.subplot(1, 3, 2)  # Create a subplot for the box plot
plt.boxplot(data5['Valor real'], vert=False)
plt.title('Box Plot Eficacia Proceso GOM')
plt.xlabel('Valor Real')

summary5.loc['mean'] = summary5.loc['mean'].round(2)
summary5.loc['std'] = summary5.loc['std'].round(2)
summary5.loc['25%'] = summary5.loc['25%'].round(2)

plt.subplot(1, 3, 3)  # Create a subplot for the summary
plt.table(cellText=summary5.values,
          rowLabels=summary5.index,
          colLabels=summary5.columns,
          loc='center')
plt.axis('off')

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# Plot the time series
plt.figure(figsize=(10, 5))
plt.plot(data6['Mes'], data5['Valor real'], label='Eficacia Proc. GOM')

# Plot the horizontal line for the limit value
plt.axhline(y=84, color='r', linestyle='--', label='Meta Actual.')
plt.axhline(y=93, color='g', linestyle='--', label='Meta Prop.')

# Add labels and legend
plt.xlabel('Mes')
plt.ylabel('Valor real')
plt.title('Eficacia Proceso GOM 2022-2024')
plt.legend()

# Set y-axis limits
plt.ylim(50, 100)

# Show the plot
plt.show()
```
